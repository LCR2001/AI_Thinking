```python
import pandas as pd
```

"stock price.xlsx" 은 종목코드(id), 회사명(name), 시가총액(value), 주가(price) 등 여러 종목의 주가 정보를 담고 있다.
"stock valuation.xlsx" 는 종목코드(id), 회사명(name), 주당순이익(eps), 주당순자산가치(bps), 주가수익비율(per), 주가자산비율(pbr) 등 주식 가치 평가 지표를 담고 있다.

```python
df_price = pd.read_excel("stock price.xlsx")
df_valuation = pd.read_excel("stock valuation.xlsx")
df_price = df_price.rename(columns = {"stock_name" : "name"})
df_stock = pd.merge(df_price, df_valuation, on =["id","name"], how ="outer")
df_stock
```

주가가 50000 이하인 종목의 이름, 주당순자산가치, 주가수익비율을 출력하라

```python
df_stock[df_stock["price"] <= 50000][["name", "bps", "per"]]
```

주가수익비율(per) 이 30 이상인 종목 이름을 출력하라. 단, 주식 가치 평가 지표가 없는 경우도 포함해야 한다.

```python
df_stock[df_stock["per"] >= 30]
```

두 데이터 프레임에 모두 존재하는 종목의 모든 관련 정보를 출력해라. 단, pd.concat()을 사용해야 한다.

```python
df_price.set_index("id", inplace = True)
df_valuation.set_index("id", inplace = True)
df_valuation.drop(columns = "name", inplace = True)
df_all = pd.concat([df_price, df_valuation], axis = 1, join = "inner")
```
