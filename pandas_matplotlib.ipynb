```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rc('font', family='NanumGothic')
```

```python
df = pd.read_excel("시도별 전출입 인구수.xlsx", engine='openpyxl', header=0, na_values="-")
df = df.fillna(method='ffill')

mask = (df['전출지별']=='서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul.rename({'전입지별':'전입지'}, axis = 1, inplace=True)
df_seoul.set_index('전입지', inplace=True)

sr_one = df_seoul.loc['경기도']

plt.plot(sr_one.index , sr_one.values)
plt.plot(sr_one)
plt.title('서울 -> 경기 인구 이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.show()
```

```python
plt.style.use('dark_background')

plt.figure(figsize=(14,5))
plt.xticks(rotation='vertical')
plt.plot(sr_one.index, sr_one.values)
plt.title('Seoul to Gyeonggi')
plt.xlabel('Days')
plt.ylabel('Population')

plt.legend(labels=['Seoul to Gyeonggi'], loc='best')
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(sr_one, 'o', markersize=10)
ax2.plot(sr_one, marker='o', markerfacecolor='green', markersize=19, color='olive', linewidth=2, label= 'Seoul to Gyeonggi')
ax2.legend(loc='best')

ax1.set_ylim(50000, 800000)
ax2.set_ylim(50000,800000)

ax1.set_xtiklabels(sr_one.index, rotation=75)
ax2.set_xtiklabels(sr_one.index, rotation=75)
```

```python
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(sr_one, marker='o', markerfacecolor='orange', markersize=10, color = 'olive', linewidth=2, label='Seoul to Gyeonggi')
ax.legend(loc='best')

ax.set_ylim(50000,800000)
ax.set_xlabel('Days', size=12)
ax.set_ylabel('Move Population', size =12)
ax.set_xticklabels(sr_one.index, rotation=75)

ax.tick_params(axis="x", labelsize=10)
ax.tick_params(axis="y", labelsize=10)

plt.show()
```

```python
col_years = list(map(str, range(1970, 2018)))
df_3 = df_seoul.loc[['충청남도','경상북도','강원도'], col_years]

plt.style.use('classic')

fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)
plt.xticks(rotation = 'vertical')

ax.plot(col_years, df_3.loc['충청남도',:], marker='o', markerfacecolor='red', markersize=10, color = 'coral', linewidth=2, label='Seoul to Chungnam')
ax.plot(col_years, df_3.loc['경상북도',:], marker='o', markerfacecolor='royalblue', markersize=10, color = 'midnightblue', linewidth=2, label='Seoul to Gyeongbuk')
ax.plot(col_years, df_3.loc['강원도',:], marker='o', markerfacecolor='violet', markersize=10, color = 'blueviolet', linewidth=2, label='Seoul to Gangwon')

ax.legend(loc='best')
ax.set_title('Seoul to Chungnam, Gyeongbuk, Gangwon', size=20)

ax.set_xlabel('Days', size=12)
ax.set_ylabel('Move Population', size=12)

plt.show()
```
